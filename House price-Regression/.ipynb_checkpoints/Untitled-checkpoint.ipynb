```python
train_test_data.drop(["Id","PoolQC","MiscFeature","Alley","Fence","FireplaceQu","LotFrontage"], axis = 1, inplace = True)
```

```python
train_test_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

#make a copy of dataframe 
X_cat = train_test_data.copy()

#select only those columns which are category type=object
X_cat = train_test_data.select_dtypes(include=['object'])
print(X_cat.shape)

#make a copy of dataframe and do one-hot encoding on that dataframe
X_enc = X_cat.copy()
print(X_enc.shape)
```

```python
X_cat.head()
```

```python
train_test_enc = pd.get_dummies(X_enc)
print(train_test_enc.shape)
```

```python
train_test_enc.head()
```

### Filling missing values

```python
#fetch missing values of a column

#train_test_enc[train_test_enc["MSZoning"].isnull()]
```

```python
# Imputing missing data by the mean of each column.
train_test_enc = train_test_enc.fillna(train_test_enc.mean())
```

```python
train_test_enc.isnull().sum().sort_values(ascending=False)
```

```python
# importing linear regressionfrom sklearn

from sklearn.linear_model import LinearRegression

lreg = LinearRegression()

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(train_test_enc,data_train.SalePrice)
```

```python
#training the model

lreg.fit(x_train,y_train)

#predicting on cv

pred = lreg.predict(x_test)

#calculating mse

mse = np.mean((pred - y_test)**2)
print(mse)
```
